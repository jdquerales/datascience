```python
!pip3 install --upgrade
!pip3 install sympy 
import sys
sys.path.append('/usr/local/lib/python3.7/site-packages') 
import sympy as sy

sy.init_printing()   # the funtion sy.init_printing() will examine the resources available to it on your machine,
                     # and endeavour to format output as best as it can.
x,y = sy.symbols("x y")  # we declare x and y as symbols
D = (x+y)*sy.exp(x)*sy.cos(y) # it is unnecessary to declare D as a symbol.
D
```

```python
# We usually also need to define unknown functions like f(x,y)
f, g = sy.symbols("f g", cls=sy.Function)
f(x), f(x,y), g(y)

# sympy neither knows nor cares about the number of funtion arguments, it is the user's responsibility!!
```

```python
# Suppose that i and j take integer number, while u and v are always real:

i, j = sy.symbols("i j", integer=True)
u, v = sy.symbols("u v", real=True)
i.is_integer, j*j, (j*j).is_integer, u.is_real

# that reminds one that symbols really are class instances, i.e. they have attributes.
```

```python
# Some useful symbols in Sympy:
sy.I, sy.E, sy.pi, sy.oo
```

```python
# In Sympy expressions are inmmutable. 
# How does one find out the result of substituting specific values for, say, x and y i D?
# This can be catered for by the substitution operator: sy.subs()

D0 = D.subs(x,0)
D0pi = D.subs(y, sy.pi)
D0, D0pi, D.subs([(x,0), (y, sy.pi)])
```

```python
x + sy.Rational(1, 3), sy.Rational('0.5')*y
```

```python
# The function sy.S converts singleton expressions, given as strings, to their Symoy equivalents:

x + sy.S('1/3'), sy.S("1/2")*y
```

```python
# Similarly the function sy.sympify converts arbitrary expressions, given as strings, to their Sympy equivalents.

D_s = sy.sympify(' (x+y)*exp(x)*cos(y) ')
cosdiff = sy.sympify("cos(x)*cos(y) + sin(x)*sin(y)")
D_s, cosdiff
```

```python
# We can obtain numerical values using the function sy.evalf, which takes a single integer argument, the precision:
# Much of the functionality of sy.evalf is also available in sy.N, which is, very nearly, a duplicate.

cospi4 = cosdiff.subs([(x,sy.pi/2),(y,sy.pi/4)])
cospi4, cospi4.evalf(), cospi4.evalf(6), sy.N(cospi4, 6)
 
# That is perfectly adequate for a few selected values of the arguments, but what about if we want to
# use the expression as a function. The Sympy function sy.lambdify does precisely that.  
```

```python
import numpy as np
func = sy.lambdify((x, y), cosdiff, 'numpy')
xn = np.linspace(0, np.pi, 13)
func(xn, 0.0)          # should return np.cos(xn)
```

```python
# For the sake of completeness we introduce matrices and vectors at this point.
# Sympy implements matrices with the Matrix class, and a n-vector as a nx1 matrix.
# The constructor requires an ordered list of rows, where each row is an ordered list of elements.
# A very simple example:

M = sy.Matrix([[1,x],[y,1]])
V = sy.Matrix([[u],[v]])
M, V, M*V
```

```python
# matrix has a number of functions/methods which can be applied to matrices;
M.eigenvects(), M.T, M.det(), M.inv(), M*M.inv()   # eigenvectors, transpose, determinant, inverse     
```

```python
# A matrix M is immutable, but its contents may be changed freely, for example:

M[0,1] = u
M
```

```python
# Differentiation. There are two ways to compute first-order derivations

sy.diff(D,x), D.diff(x)

# the second version relies on the fact that D is a symbol, i.e. a class, and this class contains
# the derivative function.

# Higher derivaties:
D.diff(x,y,y), D.diff(y,2,x), sy.diff(D,y)
```

```python
# Other examples:

D_xyy = sy.Derivative(D,x,y,2)
D_xyy, D_xyy.doit()
```

```python
# Integration
# Indefinite integration is the inverse of differentiation and SymPy treats ir accordingly.
# As with all computer algebra systems, the "constant of integration" is never shown explicitly

sy.integrate(D,y), D.integrate(y), D.integrate(x,y)
```

```python
# A "lazy" integration

yD = sy.Integral(D,y)
yD, yD.doit()
```

```python
# Definite integrattion if very similar. One simply replaces the variable of integration parameter by the tuple
# (variable, lower limit, upper limit). For example:

sy.integrate(D, (y,0,sy.pi)), D.integrate((y,0,sy.pi))
```

```python
# Multiple definite integrations are handled in the obvious way:

sy.integrate(sy.exp(-x**2-y**2), (x, 0, sy.oo), (y, 0, sy.oo))
```

```python
# there is also a "lazy" way :

dint = sy.Integral(sy.exp(-x**2-y**2), (x, 0, sy.oo), (y, 0, sy.oo))
dint, dint.doit()
```

```python
# Other examples:

sy.integrate(sy.sqrt(x+sy.sqrt(x*x))/x, x)
```

```python
sy.integrate(sy.exp(-x)*sy.sin(x**2)/x, (x, 0, sy.oo))
```

```python
# If SymPy cannot integrate your expression, it will eventually, after trying all of its store of algoritms,
# return the "lazy" integral form, a mark of failure. For example:

sy.integrate(x/sy.sin(x), x)
```

```python
# Series and Limits

foo = sy.exp(sy.sin(x))

# we would consider to expand foo as a Taylor series in x. Then term-by-term integration
# should give a useful approximation. Let say we want to expand it up to the first 10 terms 
# in the Taylor series expansion

foo_ser = foo.series(x, 0, 10)

foo_ser.removeO()

foo_ser, foo_ser.integrate(x)


```

```python
# Limits are handled in a similar way:

sy.limit((foo-1-x)/x**2, x, 0)
```

```python
# At points of discontinuity, limits from below and from above will differ. By default sy.limit takes the limit from
# above, but both limits are readily available:

goo = 1/(x-1)
goo, sy.limit(goo, x, 1)

sy.limit(goo, x, 1, dir="-"), sy.limit(goo, x, 1, dir="+")
```

```python
# Equations: a equation has both a left-hand side lhs and a right-hand side rhs:

lhs = D; rhs = cosdiff

# There are two ways in SymPy to create aan equation, both given in the following snippet:

eqn1 = sy.Eq(lhs, rhs); eqn2 = lhs - rhs 

eqn1, eqn2

# the first one produces the traditional for a equation. 
```

```python
# the universal workhorse is solve. For equations in one variable: solveset. 
# For linear system of equations: linsolve.

from sympy.solvers import solveset
from sympy.solvers.solveset import linsolve

# Some examples:

solveset(4*x-3, x), solveset(3*x**3-16*x**2+23*x-6, x)





```

```python
quad = x**2 - 2*x + 1 
solveset(quad, x), sy.roots(quad)
```

```python
# simple transcendental equations are solved over the complex plane, by default
solveset(sy.exp(x)-1, x), solveset(sy.cos(x) - x, x)
```

```python
# Linear equations with more than one independent variable

Eqns = [x+2*y, 3*x+4*y-2] # as a list
linsolve(Eqns, [x,y])
```

```python
# In many cases it will be more convenient to specify the system matrix A and the right-hand-side vector b:

A = sy.Matrix([[1,2],[2,4]])
b = sy.Matrix([[1],[2]])
linsolve((A,b),[x,y])
```

```python
A_b = sy.Matrix([[1,2,1],[2,4,1]])
linsolve(A_b,[x,y])
```

```python
# More general equations

neq = [y**2 - x**2 - 2*x - 1, 3*x - y - 1]

import sympy.solvers as sys

sys.solve(neq, [x,y])
```

```python
# Equations involving square roots

sys.solve([sy.sqrt(x) - sy.sqrt(y) - 1, sy.sqrt(x+y) - 2 ], [x,y])
```

```python
# By default solve ignores the cube roots of negative numbers! However, the situation is remeadiable.

eq1 = sy.root(3*x+1, 3) - x - 1; eq2 = sy.root(3*x-1, 3) -x + 1
sys.solve(eq1, x, check=False), sys.solve(eq2, x, check=False)
```

```python
# Solving ordinary differential equations
# Its main workhorse is dsolve

from sympy import dsolve
#dsolve?

# some ode examples:

ode1 = f(x).diff(x, 2) + 4*f(x)
sol1 = dsolve(ode1, f(x))
ode1, sol1
```

```python
# Suppose we want to impose initial conditions, say f(0) = 2, f'(0) = 0

C1, C2 = sy.symbols("C1, C2")
fun = sol1.rhs
fund = fun.diff(x)
fun.subs(x,0), fund.subs(x,0)
```

```python
psol = sol1.subs([(C2,2), (C1,0)])
psol
```

```python
ode2 = sy.sin(f(x)) + (x*sy.cos(f(x)) + f(x))*f(x).diff(x)
ode2, dsolve(ode2, f(x))
```

```python
# Bernoulli type equations:
ode3 = x*f(x).diff(x) + f(x) - sy.log(x)*f(x)**2
ode3, dsolve(ode3)
```

```python
# Inhomogeneous Euler equations

ode4 = f(x).diff(x, 2)*x**2 - 4*f(x).diff(x)*x + 6*f(x) - x**3
ode4, dsolve(ode4)
```

```python
# Liouville equation
ode5 = f(x).diff(x,2) + (f(x).diff(x))**2/f(x) + f(x).diff(x)
ode5, dsolve(ode5)
```

```python
# Some equations yield only to the technique of solution by series

ode6 = x*f(x).diff(x,2) + 2*(f(x).diff(x)) + x*f(x)
ode6, dsolve(ode6)
```

```python
ode7 = [f(x).diff(x) - 2*f(x) - g(x), g(x).diff(x) - f(x) - 2*g(x)]
ode7, dsolve(ode7)
```

```python
ode8 = f(x).diff(x) - (x+f(x))**2
ode8, dsolve(ode8)
```

```python
# in ode8 the default solution was a power series expansion about x=0. However, consider the simple transformation
# g(x) = x+f(x). This transforms the differential equation into one with the independent variable absent, 
# g'(x) = 1+g^2
```

```python
ode9 = g(x).diff(x) - 1 - (g(x))**2
dsolve(ode9)
```
